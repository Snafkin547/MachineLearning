```
def plotVec(vectors):
  ax = plt.axes()

  for vec in vectors:
    ax.arrow(0, 0, *vec["vector"], head_width = 0.05, color = vec["color"], head_length = 0.1)
    plt.text(*(vec["vector"]+0.1), vec["name"])

    plt.ylim(-2,2)
    plt.xlim(-2,2)
```
