```
import pandas as pd

df = pd.read_csv("concrete_data.csv")
df.head()
```

```
df_columns = df.columns
X = df[df_columns [df_columns !='Strength']]
y = df['Strength']
```

```
X.head()
```

```
y.head()
```

```
n_cols = X.shape[1]
print(n_cols)
```

```
def regression_model():
  
    model=Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```
import statistics
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```
model = regression_model()
mse=[]

for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=50, verbose=False)
    result = model.predict(X_test)
    mse.append(mean_squared_error(y_test, result))

print("Answer for A")
print("Mean of the mean squared errors is: %.2f" % statistics.mean(mse))
print("Standard Deviation of the mean squared errors is: %.2f" % statistics.stdev(mse))
```

```
def standardization(l):
   return (l - l.mean())/l.std()
```

```
X_norm = standardization(X)
```

```
model = regression_model()
mse=[]

for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3, random_state=4)
    model.fit(X_train, y_train, epochs=50, verbose=False)
    result = model.predict(X_test)
    mse.append(mean_squared_error(y_test, result))

print("Answer for B")
print("Mean of the mean squared errors is: %.2f" % statistics.mean(mse))
print("Standard Deviation of the mean squared errors is: %.2f" % statistics.stdev(mse))
```
